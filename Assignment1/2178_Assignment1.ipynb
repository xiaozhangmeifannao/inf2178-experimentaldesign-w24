```python

def get_summary_statistics(dataset):

    mean = np.round(np.mean(dataset), 2)
    median = np.round(np.median(dataset), 2)
    min_value = np.round(dataset.min(), 2)
    max_value = np.round(dataset.max(), 2)
    quartile_1 = np.round(dataset.quantile(0.25), 2)
    quartile_3 = np.round(dataset.quantile(0.75), 2)

    # Interquartile range
    iqr = np.round(quartile_3 - quartile_1, 2)

    print('Min: %s' % min_value)
    print('Mean: %s' % mean)
    print('Max: %s' % max_value)
    print('25th percentile: %s' % quartile_1)
    print('Median: %s' % median)
    print('75th percentile: %s' % quartile_3)
    print('Interquartile range (IQR): %s' % iqr)
```

```python
get_summary_statistics(df["SERVICE_USER_COUNT"])
```

The average number of users of shelters is 45.73, while the median number of users is only 28, indicating that small shelters still predominate and that there are a small number of shelters with a particularly high capacity. We can see the max number is 339 which should be conuted as extreme value.

```python
get_summary_statistics(df["CAPACITY_ACTUAL_BED"])
```

```python
get_summary_statistics(df["OCCUPIED_BEDS"])
```

The values for actual and occupied beds in the shelters are very close to each other, with an average difference of less than 2, and each shelter appears to be close to full capacity. Excluding the possibility that the homeless left on their own in the middle of the night, it may be that the shelters had to reserve a bed for a homeless person who might freeze to death during the night. This suggests that more shelters may need to be built to meet the large demand.

```python
grouped_df = df.groupby('PROGRAM_MODEL').agg({
    'CAPACITY_ACTUAL_BED': 'mean',
    'OCCUPIED_BEDS': 'mean'
}).reset_index()

# Plotting
labels = ['Emergency', 'Transitional']
values_capacity = grouped_df['CAPACITY_ACTUAL_BED'].values
values_occupied = grouped_df['OCCUPIED_BEDS'].values

plt.figure(figsize=(10, 6))
bar_width = 0.35
index = range(len(labels))

plt.bar(index, values_capacity, bar_width, label='Capacity', color='blue', alpha=0.7)
plt.bar([i + bar_width for i in index], values_occupied, bar_width, label='Occupied', color='orange', alpha=0.7)

plt.xlabel('Program Model')
plt.ylabel('Average Bed Count')
plt.title('Average CAPACITY_ACTUAL_BED and OCCUPIED_BEDS for Emergency and Transitional Shelters')
plt.xticks([i + bar_width/2 for i in index], labels)
plt.legend()
plt.show()
```

```python
df_cleaned = df.dropna(subset=['CAPACITY_ACTUAL_BED'])
```

```python
from scipy.stats import ttest_ind


emergency_data = df_cleaned[df_cleaned['PROGRAM_MODEL'] == 'Emergency']
transitional_data = df_cleaned[df_cleaned['PROGRAM_MODEL'] == 'Transitional']

# Extract CAPACITY_ACTUAL_BED and OCCUPIED_BEDS data for "Emergency"
capacity_emergency = emergency_data['CAPACITY_ACTUAL_BED']
occupied_emergency = emergency_data['OCCUPIED_BEDS']
capacity_transitional = transitional_data['CAPACITY_ACTUAL_BED']
occupied_transitional = transitional_data['OCCUPIED_BEDS']

# Perform two-sample t-test
t_stat_capacity, p_value_capacity = ttest_ind(capacity_emergency, capacity_transitional)
t_stat_occupied, p_value_occupied = ttest_ind(occupied_emergency, occupied_transitional)

# Print the results
print("Two-sample t-test for CAPACITY_ACTUAL_BED:")
print(f"T-statistic: {t_stat_capacity}")
print(f"P-value: {p_value_capacity}")

print("\nTwo-sample t-test for OCCUPIED_BEDS:")
print(f"T-statistic: {t_stat_occupied}")
print(f"P-value: {p_value_occupied}")
```

```python
# Now lets see if the utilization_rate are different for two program method
emergency_utilization_rate = occupied_emergency / capacity_emergency
transitional_utilization_rate = occupied_transitional / capacity_transitional
t_stat_utilization_rate, p_stat_utilization_rate = ttest_ind(emergency_utilization_rate, transitional_utilization_rate)


# Print the results
print("Two-sample t-test for utilization_rate:")
print(f"T-statistic: {t_stat_utilization_rate}")
print(f"P-value: {p_stat_utilization_rate}")
```
